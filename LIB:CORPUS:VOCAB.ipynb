```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python

def tokenize_collection(df):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    
    all_tokens = []

    for index, row in df.iterrows():
        content = row['doc_content']
        doc_id = row['doc_id']
        tokens = word_tokenize(content)
        pos_tags = pos_tag(tokens)

        for token, tag in pos_tags:
            if token.lower() not in stop_words and token.isalnum():
                lemma = lemmatizer.lemmatize(token)
                pos_group = get_pos_group(tag)  # You need to define this function
                all_tokens.append((doc_id, token, lemma, tag, pos_group))

    tokens_df = pd.DataFrame(all_tokens, columns=['doc_id', 'token_str', 'term_str', 'pos', 'pos_group'])
    tokens_df.set_index(['doc_id', 'token_str'], inplace=True)
    
    return tokens_df

def get_pos_group(tag):
    if tag.startswith('N'):
        return 'Noun'
    elif tag.startswith('V'):
        return 'Verb'
    elif tag.startswith('J'):
        return 'Adjective'
    elif tag.startswith('R'):
        return 'Adverb'
    else:
        return 'Other'
```

```python
CORPUS = tokenize_collection(df)
print(CORPUS)
```

```python
VOCAB['max_pos'] = CORPUSS[['term_str','pos']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB['max_pos_group'] = CORPUSS[['term_str','pos_group']].value_counts().unstack(fill_value=0).idxmax(1)
VOCAB
```

```python
#computing POS ambiguity
VOCAB['n_pos_group'] = CORPUSS[['term_str','pos_group']].value_counts().unstack().count(1)
VOCAB['cat_pos_group'] = CORPUSS[['term_str','pos_group']].value_counts().to_frame('n').reset_index()\
    .groupby('term_str').pos_group.apply(lambda x: set(x))
VOCAB['n_pos'] = CORPUSS[['term_str','pos']].value_counts().unstack().count(1)
VOCAB['cat_pos'] = CORPUSS[['term_str','pos']].value_counts().to_frame('n').reset_index()\
    .groupby('term_str').pos.apply(lambda x: set(x))
VOCAB
```
